```python
import pandas as pd
import json
```

**Match Results**

```python
with open('t20_wc22_json_files/t20_wc_match_results.json') as f:
    data = json.load(f)
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard': 'match_id'}, axis = 1, inplace = True)
df_match.head()
```

```python
''' 
creating dict like this....
{
    "Namibia Vs Sri Lanka" : "T20I # 1823",
    "Sri Lanka Vs Namibia" : "T20I # 1823",
    "Netherlands Vs U.A.E." : "T20I # 1825",
    "U.A.E. Vs Netherlands" : T20I # 1825,
}
'''

match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']    

# match_ids_dict
```

**Batting Summary**

```python
with open('t20_wc22_json_files/t20_wc_batting_summary.json') as f:
    data = json.load(f)
    
    all_matches_record = []
    
    for rec in data:
        all_matches_record.extend(rec['battingSummary'])
        
df_bat = pd.DataFrame(all_matches_record)
df_bat.tail(11)
```

```python
df_bat.shape
```

```python
df_bat["out/not_out"] = df_bat.dismissal.apply(lambda x: "out" if len(x) > 0 else "not_out")
df_bat.tail(11)
```

```python
df_bat.drop(columns = ["dismissal"], inplace = True)
```

```python
df_bat.tail(11)
```

```python
df_bat["batsmanName"] = df_bat["batsmanName"].apply(lambda x: x.replace('â€',''))
df_bat["batsmanName"] = df_bat["batsmanName"].apply(lambda x: x.replace('\xa0',''))
df_bat.head(30)
```

```python
df_bat["match_id"] = df_bat["match"].map(match_ids_dict)

df_bat.head(30)
```

```python
df_bat.to_csv('t20_csv_files/fact_batting_summary.csv', index = False)
```

***Bowling summary***

```python
with open('t20_wc22_json_files/t20_wc_bowling_summary.json') as f:
    data = json.load(f)
    all_record = []
    for rec in data:
        all_record.extend(rec["bowlingSummary"])

df_bowling = pd.DataFrame(all_record)
df_bowling.head(11)
```

```python
df_bowling.shape
```

```python
df_bowling["match_id"] = df_bowling["match"].map(match_ids_dict)
df_bowling.head(11)
```

```python
df_bowling.to_csv('t20_csv_files/fact_bowling_summary.csv', index = False)
```

***Process Players Information***

```python
with open('t20_wc22_json_files/t20_wc_player_info.json') as f:
    data = json.load(f)

df_player_info = pd.DataFrame(data)
print(df_player_info.shape)
df_player_info.head(11)
```

```python
# cleanup wired characters

df_player_info['name'] = df_player_info['name'].apply(lambda x: x.replace('â€', ''))
df_player_info['name'] = df_player_info['name'].apply(lambda x: x.replace('†', ''))
df_player_info['name'] = df_player_info['name'].apply(lambda x: x.replace('\xa0', ''))
df_player_info.tail(11)
```
